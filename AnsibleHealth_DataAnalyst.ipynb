```python
columns = ['Name', 'Gender', 'Count']
df_2007 = pd.read_csv('yob2007.txt', names=columns)
df_2020 = pd.read_csv('yob2020.txt', names=columns)
```

```python
top_male_2007 = df_2007[df_2007['Gender'] == 'M'].sort_values(by='Count', ascending=False).head(10)
top_female_2007 = df_2007[df_2007['Gender'] == 'F'].sort_values(by='Count', ascending=False).head(10)

print("Top 10 male names in 2007:\n", top_male_2007)
print("\nTop 10 female names in 2007:\n", top_female_2007)
```

```python
top_male_2020 = df_2020[df_2020['Gender'] == 'M'].sort_values(by='Count', ascending=False).head(100)
top_female_2020 = df_2020[df_2020['Gender'] == 'F'].sort_values(by='Count', ascending=False).head(100)

print("Top 10 male names in 2020:\n", top_male_2020)
print("\nTop 10 female names in 2020:\n", top_female_2020)
```

```python
# Top 100 names in 2007 instead of top 10
top_male_2007 = df_2007[df_2007['Gender']=='M'].sort_values(by='Count', ascending=False).head(100)
top_female_2007 = df_2007[df_2007['Gender']=='F'].sort_values(by='Count', ascending=False).head(100)
```

```python
# Names from 2007 top 100 that are NOT in 2020 top 100
dropped_male_names = top_male_2007[~top_male_2007['Name'].isin(top_male_2020['Name'])]
dropped_female_names = top_female_2007[~top_female_2007['Name'].isin(top_female_2020['Name'])]

print("Dropped Male Names:\n", dropped_male_names.head(10))  # show top 10 dropped
print("\nDropped Female Names:\n", dropped_female_names.head(10))
```
